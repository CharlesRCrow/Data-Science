# Finding Best Two Markets to Advertise In

Hypothetically a course for e-learning has been created.  There is a limited budget on advertising so need to determine what the best market is to advertise in.  [New Coder Survey](https://medium.freecodecamp.org/we-asked-20-000-people-who-they-are-and-how-theyre-learning-to-code-fff5d668969)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

survey=pd.read_csv('2017-fCC-New-Coders-Survey-Data.csv', low_memory=False)
print(survey.head())
```

# Checking for Sample Representativity
**Determine how representative the sample is based on job interest, age, location, and money willing to be spent.**

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10,10))
colors = ['#ff9999','#66b3ff']
survey['JobRoleInterest'].notnull().value_counts().plot(kind='pie',y='Current Job Interest', autopct='%1.0f%%', colors=colors, legend=True, explode=[0.05, 0.05], radius=1.0)
plt.title('Stated Specific Job Interest', weight='bold', size=18)
plt.yticks(fontsize=18, weight='bold')
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left', fontsize=18)
plt.xlabel('')
plt.ylabel('')
```

**Only 38% of survey takers stated a current job interest**

```python
plt.figure(figsize=(10,8))

survey['JobRoleInterest'].value_counts(ascending=False).head(10).plot.barh()

plt.title('Top 10 Job Interests', weight='bold', size=18)
plt.show()
```

```python
plt.figure(figsize=(10,8))

interests_no_nulls = survey['JobRoleInterest'].dropna()
web_or_mobile = interests_no_nulls.str.contains(
    'Web Developer|Mobile Developer') 
freq_table = web_or_mobile.value_counts(normalize = True) * 100

freq_table.plot.bar()
plt.title('Most Participants are Interested in \nWeb or Mobile Development', weight='bold', size=18)
plt.ylabel('Percentage', fontsize = 18, weight='bold')
plt.xticks([0,1],['Web or mobile\ndevelopment', 'Other subject'],
           rotation = 0, weight='bold', size=18) 
plt.ylim([0,100])
plt.show()
```

**Over 80% are interested with web or mobile development.  Full-Stack Web Developer is the most popular.**

# Age Profile of Survey Takers

**First we will analyse the age profile of people learning to code.**

```python
survey['Age'].describe()
```

```python
intervals = pd.interval_range(start = 0, end = 90, freq = 10)
gr_freq_table = pd.Series([0,0,0,0,0,0,0,0,0], index = intervals)

for value in survey['Age']:
    for interval in intervals:
        if value in interval:
            gr_freq_table.loc[interval] += 1
            break
under_10=11/15367*100
print(gr_freq_table)
print('% less than 10 years old: ', under_10)
```

The data looks fairly reasonable.  A few senior citizens and childen learning to program is not unlikely.  There are likely a few typos as it is unlikely that an infant a few months old will be filling out a survey.  

```python
plt.figure(figsize=(10,6))
survey['Age'].plot.hist(bins=10)
plt.title('Age Profile', weight='bold', size=18)
plt.axvline(survey['Age'].mean(), label='mean', color='Green')
plt.axvline(survey['Age'].median(), label='median', color='Purple')
plt.axvline(25, label='mode', color='Red')
plt.xlabel('Years Old')
plt.legend()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(survey['Age'])
plt.title('Age Profile', weight='bold', size=18)
```

**The 20-30 age bracket is the largest age demographic that needs to be appealed to.  The 25%-75% quartiles range is 22-32 years old.**

# New Coders - Locations and Densities

```python
survey_good = survey[survey['JobRoleInterest'].notnull()].copy()
country=survey_good['CountryLive']
country_freq=country.value_counts(ascending=False)
plt.figure(figsize=(10,8))

country_freq.head(10).plot.barh()

plt.title('Top 10 Countries in Survey', weight='bold', size=22)
plt.show()
```

**USA is by far has the most respondents.  India has about 1/6 the amount of responses to the survey vs the US.**

```python
plt.figure(figsize=(10,10))
colors = ['#ff9999','#66b3ff']
mask=survey_good['CountryLive']=='United States of America'

t_f=mask.value_counts()
label=['Other', 'USA']
plt.pie(t_f, labels=label,autopct='%1.0f%%',colors=colors,explode=[0.05, 0.05])
# mask.value_count().plot(kind='pie',y='Current Job Interest', autopct='%1.0f%%', colors=colors, legend=True)
      
plt.title('Current Country of Respondents', weight='bold', size=22)
plt.xticks(size=18, weight='bold')
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left', fontsize=18)
plt.xlabel('')
plt.ylabel('')
```

**45% of programmers taking the survey are from the United States of America.**

# Distribution of Months Learning

```python
#Rounding mounths learning from 0 to 1 month to make math easier later on and better reflect on experience.
survey_good['MonthsProgramming'].replace(0,1, inplace=True)

plt.figure(figsize=(10,6))
sns.violinplot(survey_good['MonthsProgramming'])
plt.title('Distribution of Months Learning by New Coders', weight='bold', size=22)
plt.xlim(-20, 100)
plt.axvline(survey['MonthsProgramming'].mean(), label='mean', color='Green')
plt.axvline(survey['MonthsProgramming'].median(), label='median', color='Purple')
plt.legend()
```

**The average amount of month spent learning for the respondents is about 20 months with the median being just a bit under 12 months.**

# Spending Money for Learning

```python
survey_good['spent_per_month']=survey_good['MoneyForLearning']/survey_good['MonthsProgramming']

countries_mean = survey_good.groupby('CountryLive').mean()

freq_spent=countries_mean['spent_per_month'][['United States of America','India', 'United Kingdom','Canada']]
plt.figure(figsize=(10,6))
freq_spent.plot.barh()
plt.ylabel('')
plt.title('Cash Spent Per Month on Learning to Code', weight='bold', size=18)
```

**It looks odd that that India spends more per capita than both Canada and the UK.  This could be because the data is not representative or there are huge amounts of outliers.**

```python
survey_good = survey_good[survey_good['CountryLive'].notnull()]

country_4=survey_good[survey_good['CountryLive'].str.contains('United States of America|India|United Kingdom|Canada')]
plt.figure(figsize=(10,6))
sns.boxplot(data=country_4, x='spent_per_month', y='CountryLive')
```

**There are two datapoints above 20,000 per month.  While it is possible for bootcamps to be between 5,000 to 20,000 dollars, it is unlikely to cost over 20,000 per month.  In all likelihood, it would be better to discount any values over 20,000.**

```python
country_4=country_4[country_4['spent_per_month']<20000]
country_4_mean = country_4.groupby('CountryLive').mean()

plt.figure(figsize=(10,6))
country_4_mean['spent_per_month'].plot.barh()
plt.ylabel('')
plt.title('Cash Spent Per Month on Learning to Code', weight='bold', size=18)
```

**Dropping the extreme outliers has dropped the average cash spent per month in United States about 40 dollars.**
