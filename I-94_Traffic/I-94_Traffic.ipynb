Looking to discover what factors led to heavy traffic (greater than 5000 vehicles per hour) on I-94.
Station recording the data is midway between Minneapolis and Saint Paul and records on westbound traffic.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

metrotraffic=pd.read_csv('Metro_Interstate_Traffic_Volume.csv')
metrotraffic.info()
```

Creating histogram to analyze traffic volume.

```python
metrotraffic['traffic_volume'].plot.hist()
metrotraffic['traffic_volume'].describe()
```

Between 2012-10-02 09:00:00	and 2018-09-30 23:00:00	the traffic varied between 0 and 7280 vehicles per hour.  50% of the time the traffic was between 3380 and 7280 vehicles per hour.  This high level of traffic was likely during daytime hours.  During the slowest 25% interval the traffic was below 1193 vehicles per hour likely during the slowest nightime hours.

# Traffic Volume: Day vs. Night

```python
metrotraffic['date_time'] = pd.to_datetime(metrotraffic['date_time'])

day = metrotraffic.copy()[(metrotraffic['date_time'].dt.hour >= 7) & (metrotraffic['date_time'].dt.hour < 19)]
print(day.shape)

night = metrotraffic.copy()[(metrotraffic['date_time'].dt.hour >= 19) | (metrotraffic['date_time'].dt.hour < 7)]
print(night.shape)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
day['traffic_volume'].plot.hist()
plt.title('Day Traffic')
plt.xlabel('Traffic Volume')
plt.ylim(0,8000)

plt.subplot(1, 2, 2)
night['traffic_volume'].plot.hist()
plt.title('Night Traffic')
plt.xlabel('Traffic Volume')
plt.ylim(0,8000)
plt.xlim(0,7000)
plt.show()
```

```python
day['traffic_volume'].describe()
```

```python
night['traffic_volume'].describe()
```

The Day Traffic data skews to the right with higher traffic volume with an average of 4762 vehicles per hour.  The Night Traffic data skes to the left has an overall lower traffic volume with an average of 1785 vehicles per hour.  While day traffic is about 250% more than night traffic, there are still a few significant spikes of traffic volume during nightime hours.  This could be due to holiday travel or emergencies.  We will focus on data recorded during the day.

# Time Indicators by Month

```python
day['month']=day['date_time'].dt.month
by_month=day.groupby('month').mean()
by_month['traffic_volume']
```

```python
plt.plot(by_month['traffic_volume'])
```

There is a drop in traffic in July and December.  Likely due to holidays and school being out for summer/winter breaks.

# Time Indicators by Day of Week

```python
day['dayofweek'] = day['date_time'].dt.dayofweek
by_dayofweek = day.groupby('dayofweek').mean()
by_dayofweek['traffic_volume'].plot()
```

0 is Monday and 6 is Sunday.  Traffic falls off during the weekend and doesn't reach its peak until Thurday and Friday.

# Time Indicators by Business Days vs Weekends

```python
day['hour'] = day['date_time'].dt.hour
bussiness_days = day.copy()[day['dayofweek'] <= 4] # 4 == Friday
weekend = day.copy()[day['dayofweek'] >= 5] # 5 == Saturday
by_hour_business = bussiness_days.groupby('hour').mean()
by_hour_weekend = weekend.groupby('hour').mean()

print(by_hour_business['traffic_volume'])
print(by_hour_weekend['traffic_volume'])
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
by_hour_business['traffic_volume'].plot.line()
plt.title('Business Traffic')
plt.xlabel('Hour')
plt.ylabel('Traffic Volume')
plt.ylim(0,7000)

plt.subplot(1, 2, 2)
by_hour_weekend['traffic_volume'].plot.line()
plt.title('Weekend Traffic')
plt.xlabel('Hour')
plt.ylim(0,7000)
plt.show()
```

Busiest traffic during the work week from about 5am to 8am and spikes again at 4pm.  Weekend traffic plateaus from 12pm to 4pm.

# Weather Indicators

```python
day.corr()['traffic_volume']
```

The Strongest corrolation with traffic volume is with temperature rise.

```python
day.plot.scatter('traffic_volume','temp')
plt.ylim(230,320)
```

There doesn't appear to be any clear corrolation between temp and traffic volume.  No other numerical column data has any corrolation to traffic volume.  We will now analyze the two columns containing categorical data.

```python
by_weather_main = day.groupby('weather_main').mean()
by_weather_description = day.groupby('weather_description').mean()

by_weather_main['traffic_volume'].plot.barh()
plt.show()
```

None of these exceed 5000 vehicles per hour (no heavy traffic).  We will now plot by weather description.

```python
by_weather_description['traffic_volume'].plot.barh(figsize=(4,9))
```

The two biggest heavy traffic indicators are **shower snow** and **light rain and snow**.  The likely hood is that these two weather patterns slow traffic on I-94 causing the vehicles to stay on the highway longer and the weather is not severe enough to persuade motorists to avoid driving.
