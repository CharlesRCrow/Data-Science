# Star Wars Survey

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
print(star_wars.head(10))
star_wars.columns
```

```python
sns.heatmap(star_wars.isnull())
plt.title('null values not shaded')
```

# Cleaning and Mapping Yes/No Columns

Convert columns to boolean type.

```python
yes_no={'Yes':True, 'No':False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?']=star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']=star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)

print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].head())
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].head()
```

# Cleaning and Mapping Checkbox Columns

```python
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for c in star_wars.columns[3:9]:
    star_wars[c] = star_wars[c].map(movie_mapping)

star_wars.head()
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_pm",
        "Unnamed: 4": "seen_aotc",
        "Unnamed: 5": "seen_rots",
        "Unnamed: 6": "seen_anh",
        "Unnamed: 7": "seen_esb",
        "Unnamed: 8": "seen_rotj"
        })

star_wars.head()
```

# Cleaning the Ranking Columns

```python
for c in star_wars.columns[9:15]:
    star_wars[c]=pd.to_numeric(star_wars[c], errors='coerce')
star_wars.info()    
```

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "rank_pm",
        "Unnamed: 10": "rank_aotc",
        "Unnamed: 11": "rank_rots",
        "Unnamed: 12": "rank_anh",
        "Unnamed: 13": "rank_esb",
        "Unnamed: 14": "rank_rotj"
        })
star_wars.head()
```

# Finding Highest-Ranked Star Wars Movie

```python
ranking_mean=star_wars[star_wars.columns[9:15]].mean()
ranking_mean
```

```python
sns.barplot(x=star_wars.columns[9:15], y=ranking_mean)
plt.title('Star Wars Ranks (lower ranking is better)', weight='bold')
```

The 'original' trilogy is ranked much higher than the later 'prequel' trilogy.  **Revenge of the Sith** ranks the lowest while **Empire Strkes Back** is ranked highest.

# Most Viewed Movie

```python
view_sum=star_wars[star_wars.columns[3:9]].sum()
view_sum
```

```python
sns.barplot(x=star_wars.columns[3:9], y=view_sum)
plt.title('Star Wars Total Views in Survey', weight='bold')
```

**Empire Strikes Back** is the most viewed movie followed closely by **Return of the Jedi**.

# Views and Ranking by Star Trek Fans

```python
trek_fan=star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"] == "Yes"]
trek_no=star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"] == "No"]
```

```python
sns.lineplot(x=star_wars.columns[9:15], y=trek_fan[trek_fan.columns[9:15]].mean(), label='Star Trek Fan', sort=False)
sns.lineplot(x=star_wars.columns[9:15], y=trek_no[trek_no.columns[9:15]].mean(), label='Non-Star Trek Fan', sort=False)
plt.title('Star Wars Movie Ranking by Star Trek Fans', weight='bold')
plt.legend()
```

Star Trek Fans are overall more likely to rank **Empire Strikes Back** as their favorite vs Non-Star Trek Fans.  Overall trends are similar between the two groups.

```python
sns.lineplot(x=star_wars.columns[3:9], y=trek_fan[trek_fan.columns[3:9]].sum(), label='Star Trek Fan', sort=False)
sns.lineplot(x=star_wars.columns[3:9], y=trek_no[trek_no.columns[3:9]].sum(), label='Non-Star Trek Fan', sort=False)
plt.title('Seen Star Wars movie by Star Trek Fans', weight='bold')
plt.legend()
```

Star Trek fans are more likely to watch Star Wars vs non-Trek fans.
